# WLS Algorithm for dataframe with one feature

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

years = np.arange(1980, 2020)
GDP_values = np.array([
    np.nan, np.nan, np.nan, 2.1, 2.4, 2.7, 3.0, 3.5, 4.0, 4.5, 
    5.0, 5.6, 6.2, 7.0, 8.0, 9.0, 10.5, 11.5, 12.5, 13.0, 
    13.8, 14.2, 14.5, 14.8, 15.0, 15.2, 15.3, 15.5, 16.0, 16.5, 
    17.0, 17.2, 17.5, 17.7, 18.0, 18.2, 18.4, 18.6, 18.8, 19.0
])

df = pd.DataFrame({'GDP': GDP_values}, index=years)


valid_data = df.dropna()  
missing_data = df[df['GDP'].isna()]  



X_valid = valid_data.index.values.reshape(-1, 1)  
y_valid = valid_data['GDP'].values  
X_missing = missing_data.index.values.reshape(-1, 1)  

weights = 1 / (np.abs(X_valid - X_missing.T) + 1) ** 2  
X_valid_wls = sm.add_constant(X_valid)  
model = sm.WLS(y_valid, X_valid_wls, weights.mean(axis=1))  
results = model.fit()

X_missing_wls = sm.add_constant(X_missing)  
predicted_GDP = results.predict(X_missing_wls)


df.loc[df['GDP'].isna(), 'GDP'] = predicted_GDP  
```

# WLS Algorithm for dataframe with more than one feature

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

years = np.arange(1980, 2020)
data = {
    'GDP': [np.nan, np.nan, np.nan, 2.1, 2.4, 2.7, 3.0, 3.5, 4.0, 4.5, 
            5.0, 5.6, 6.2, 7.0, 8.0, 9.0, 10.5, 11.5, 12.5, 13.0, 
            13.8, 14.2, 14.5, 14.8, 15.0, 15.2, 15.3, 15.5, 16.0, 16.5, 
            17.0, 17.2, 17.5, 17.7, 18.0, 18.2, 18.4, 18.6, 18.8, 19.0],
    
    'Population': [np.nan, np.nan, np.nan, 50, 52, 55, 57, 60, 64, 68, 
                   72, 75, 80, 85, 90, 95, 100, 110, 120, 125, 
                   130, 135, 140, 145, 150, 155, 160, 165, 170, 175, 
                   180, 182, 185, 187, 190, 192, 194, 196, 198, 200],

    'Export': [np.nan, np.nan, np.nan, 0.5, 0.55, 0.6, 0.7, 0.8, 0.9, 1.0, 
               1.1, 1.2, 1.3, 1.4, 1.5, 1.7, 1.9, 2.0, 2.2, 2.3, 
               2.4, 2.5, 2.6, 2.7, 2.8, 3.0, 3.2, 3.3, 3.5, 3.7, 
               3.9, 4.0, 4.2, 4.3, 4.5, 4.7, 4.9, 5.0, 5.2, 5.4]
}

df = pd.DataFrame(data, index=years)


for column in df.columns:
    valid_data = df.dropna(subset=[column])  
    missing_data = df[df[column].isna()]  
    
    if missing_data.empty:
        continue  

    X_valid = valid_data.index.values.reshape(-1, 1)  
    y_valid = valid_data[column].values  
    X_missing = missing_data.index.values.reshape(-1, 1)  

    weights = 1 / (np.abs(X_valid - X_missing.T) + 1) ** 2

    
    X_valid_wls = sm.add_constant(X_valid)
    model = sm.WLS(y_valid, X_valid_wls, weights.mean(axis=1))
    results = model.fit()


    X_missing_wls = sm.add_constant(X_missing)
    predicted_values = results.predict(X_missing_wls)

    df.loc[df[column].isna(), column] = predicted_values
```

# WLS Algorithm for dataframe with more than one feature and more than one country

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm


data = {
    'Country': ['USA', 'USA', 'USA', 'USA', 'USA', 'FRA', 'FRA', 'FRA', 'FRA', 'FRA'],
    'Year': [2000, 2001, 2002, 2003, 2004, 2000, 2001, 2002, 2003, 2004],
    'GDP': [np.nan, np.nan, np.nan, 9.9, 10.0, np.nan, 11.5, 2.0, 2.5, 3.0],
    'Population': [np.nan, np.nan, 290, 300, 305, np.nan, 315, 60, 63, 65]
}

df = pd.DataFrame(data)
df.set_index(['Country', 'Year'], inplace=True)  


for country in df.index.get_level_values(0).unique():  
    country_df = df.loc[country]  

    for column in df.columns:  
        valid_data = country_df.dropna(subset=[column])  
        missing_data = country_df[country_df[column].isna()]  

        if missing_data.empty:
            continue  

        X_valid = valid_data.index.values.reshape(-1, 1)  
        y_valid = valid_data[column].values  
        X_missing = missing_data.index.values.reshape(-1, 1) 
    
        X_valid_wls = sm.add_constant(X_valid)

        
        weights = 1 / (np.abs(X_valid - X_missing.T) + 1) ** 2  
        
        model = sm.WLS(y_valid, X_valid_wls, weights.mean(axis=1))
        results = model.fit()

        
        X_missing_wls = np.ones((X_missing.shape[0], X_valid_wls.shape[1]))  
        X_missing_wls[:, -1] = X_missing.ravel()  

        predicted_values = results.predict(X_missing_wls)

        df.loc[(country, missing_data.index), column] = predicted_values
```
