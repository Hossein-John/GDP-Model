# Reshape df_inflation : Melt to "Country", "Year", and "Value" format

```python
df_inflation
```

```python
df_inflation_long = df_inflation.reset_index().melt(
    id_vars=["Country Code"], 
    var_name="Year",  
    value_name="Inflation" 
)
```

```python
df_inflation_long
```

```python
df_inflation_long = df_inflation_long.sort_values(by=["Country Code", "Year"])
```

```python
df_inflation_long.set_index(['Country Code', 'Year'], inplace = True)
```

```python
df_inflation_long
```

```python
df_inflation_long.loc[("ABW", '2020')]
```

```python
df_inflation_long.to_csv("inflation_preprocessed.csv")
```
