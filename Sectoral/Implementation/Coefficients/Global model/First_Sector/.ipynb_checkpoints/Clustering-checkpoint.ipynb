```python
data = pd.read_csv('Data.csv', index_col = [0,1])
data
```

```python
df_firstSector = data.drop(columns = ['Second_sector_GDP', 'Third_sector_GDP', 'total_GDP', 'total_Investment', 'Second_sector_Inv', 'Third_sector_Inv'])
```

```python
from numpy.linalg import eigvals
from sklearn.preprocessing import StandardScaler 
import matplotlib.pyplot as plt

countries = df.index.get_level_values(0).unique()
features = ['ED', 'Expenditure', 'Irate', 'Population', 'Tax', 'Trade', 'First_sector_Inv']
results = []

for country in countries:
    df_country = df.loc[country, features]
    
    X_scaled = StandardScaler().fit_transform(df_country)
    cov_matrix = np.dot(X_scaled.T, X_scaled)
    eigenvalues = eigvals(cov_matrix)
    condition_number = max(eigenvalues) / min(eigenvalues)

    results.append({
        'Country': country, 
        'Min Eigenvalue': round(min(eigenvalues), 4),
        'Max Eigenvalue': round(max(eigenvalues), 4),
        'Condition Number': round(condition_number, 2)
    })

results_df = pd.DataFrame(results)
```

```python
import numpy as np
from numpy.linalg import eig
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Example: Brazil
df_country = df.loc['Spain', features].dropna()
X_scaled = StandardScaler().fit_transform(df_country)

# Step 1: Covariance matrix
cov_matrix = np.dot(X_scaled.T, X_scaled)

# Step 2: Get eigenvalues and eigenvectors
eigenvalues, eigenvectors = eig(cov_matrix)

# Step 3: Sort by smallest eigenvalue
idx = np.argsort(eigenvalues)  # index of sorted eigenvalues
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]

# Step 4: Show the eigenvector with the smallest eigenvalue
smallest_eigval = eigenvalues[0]
critical_eigenvector = eigenvectors[:, 0]

# Step 5: Combine with feature names for interpretation
contribution = pd.Series(critical_eigenvector, index=features)
contribution = contribution.sort_values(key=abs, ascending=False)

print("🔍 Smallest Eigenvalue:", round(smallest_eigval, 6))
print("\n🚨 Features contributing to multicollinearity:")
print(contribution)
```

```python
import numpy as np
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# انتخاب ویژگی‌ها برای کلاسترینگ
features = ['ED', 'Expenditure', 'Irate', 'Population', 'Tax', 'Trade', 'First_sector_Inv']

# محاسبه میانگین هر ویژگی برای هر کشور (برای ایجاد یک نمایه مشخص برای هر کشور)
country_features = df[features].groupby(level=0).mean()

# استانداردسازی ویژگی‌ها
X_scaled = StandardScaler().fit_transform(country_features)

# انجام کلاسترینگ سلسله‌مراتبی (روش Single Linkage)
Z = linkage(X_scaled, method='single')

# رسم دندروگرام
plt.figure(figsize=(12, 6))
dendrogram(Z, labels=country_features.index, leaf_rotation=90)
plt.title('Hierarchical Clustering of Countries (Single Linkage)')
plt.xlabel('Country')
plt.ylabel('Distance')
plt.tight_layout()
plt.show()
```

# DBSCAN clustering

```python
from sklearn.cluster import DBSCAN

features = ['ED', 'Expenditure', 'Irate', 'Population', 'Tax', 'Trade', 'First_sector_Inv']

country_features = df[features].groupby(level = 0).mean()

X_scaled = StandardScaler().fit_transform(country_features)

dbscan = DBSCAN(min_samples=2, eps = 1)

dbs_clusters = dbscan.fit_predict(X_scaled)
```

```python
dbs_clusters
```

```python
country_features['DBSCAN_Cluster'] = dbs_clusters
```

```python
country_features['DBSCAN_Cluster'].value_counts()
```

```python
country_features['DBSCAN_Cluster'] = dbs_clusters
country_features.to_csv('DBSCAN_Clusters.csv')
df_1 = country_features
```

# PCA 2 Dimentions

```python
from sklearn.decomposition import PCA

features = ['ED', 'Expenditure', 'Irate', 'Population', 'Tax', 'Trade', 'First_sector_Inv']

country_features = df[features].groupby(level = 0).mean()

X_scaled = StandardScaler().fit_transform(country_features)

pca = PCA(n_components = 0.79, random_state=42)

X_pca = pca.fit_transform(X_scaled)
```

```python
print("Number of components:", pca.n_components_)
```

```python
pca_df = pd.DataFrame(X_pca, index = country_features.index, columns = ['component_1', 'component_2'])
```

```python
pca_df
```

```python
pca_df['DBSCAN_cluster'] = dbs_clusters
```

```python
pca_df
```

```python
fig = plt.figure(figsize = (10,6))
scatter = plt.scatter(
    pca_df['component_1'], pca_df['component_2'], c = pca_df['DBSCAN_cluster'], cmap = 'Set1', s = 80, edgecolor = 'black',alpha = 0.9)

plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('DBSCAN Clusters (PCA-reduced 2D)')
fig.colorbar(scatter, label = 'Cluster Label')
plt.grid()
plt.show()
```
