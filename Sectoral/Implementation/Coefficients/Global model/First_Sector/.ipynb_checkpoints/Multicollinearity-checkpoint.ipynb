```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm

# List of features you want to check
features = ['ED', 'Expenditure', 'Irate', 'Population', 'Tax', 'Trade', 'First_sector_GDP', 'First_sector_Inv']

# Store results
vif_by_country = {}

for country, group in grouped:
    X = group[features]
    
    vif = pd.DataFrame()
    vif["Feature"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    
    vif_by_country[country] = vif
```

```python
print(vif_by_country['Luxembourg'])
```

```python
summary = pd.DataFrame({
    country: vif.set_index("Feature")["VIF"] 
    for country, vif in vif_by_country.items()
})
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import RidgeCV, Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# === Step 1: Define your target and predictors ===
target = 'First_sector_GDP'
features = ['ED', 'Expenditure', 'Irate', 'Population', 'Tax', 'Trade', 'First_sector_Inv']

# === Step 2: Define alpha range to search ===
alphas = np.logspace(-4, 4, 100)

# === Step 3: Initialize results dictionary ===
results = {}

# === Step 4: Loop through each country ===
for country, group in df.groupby(level=0):  # df must be a MultiIndex with (Country, Year)
    print(f"\n🔍 Processing country: {country}")
    
    group = group.dropna()

    if len(group) < len(features) + 1:
        print("⚠️ Not enough data to train the model.")
        continue

    X = group[features].copy()
    y = group[target]

    # === Step 5: Scale predictors ===
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # === Step 6: Split into train and test sets ===
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

    # === Step 7: RidgeCV to find best alpha ===
    ridge_cv = RidgeCV(alphas=alphas, store_cv_values=True)
    ridge_cv.fit(X_train, y_train)
    best_alpha = ridge_cv.alpha_

    # === Step 8: Fit final Ridge model with best alpha ===
    final_model = Ridge(alpha=best_alpha)
    final_model.fit(X_train, y_train)

    # === Step 9: Make predictions ===
    y_train_pred = final_model.predict(X_train)
    y_test_pred = final_model.predict(X_test)

    # === Step 10: Calculate metrics ===
    train_r2 = final_model.score(X_train, y_train)
    test_r2 = final_model.score(X_test, y_test)
    train_mse = mean_squared_error(y_train, y_train_pred)
    test_mse = mean_squared_error(y_test, y_test_pred)

    # === Step 11: Store results ===
    results[country] = {
        "Best Alpha": best_alpha,
        "Train R²": train_r2,
        "Test R²": test_r2,
        "Train MSE": train_mse,
        "Test MSE": test_mse
    }

# === Step 12: Convert results to a summary DataFrame ===
summary_df = pd.DataFrame.from_dict(results, orient='index').reset_index()
summary_df = summary_df.rename(columns={'index': 'Country'})

# === Step 13: Show the summary (optional: save it) ===
print("\n📊 Ridge Regression Summary:")
print(summary_df)

# Optional: Save the summary to a CSV file
# summary_df.to_csv("ridge_summary_per_country.csv", index=False)
```
