# Interest rate

```python
df_Irate = df_Irate.iloc[:,::2]
```

```python
df_Irate = df_Irate.drop('GEO (Labels)')
```

```python
eurozone_countries = [
    'Austria', 'Belgium', 'Croatia', 'Cyprus', 'Estonia', 'Finland', 
    'France', 'Germany', 'Greece', 'Ireland', 'Italy', 'Latvia', 
    'Lithuania', 'Luxembourg', 'Malta', 'Netherlands', 'Portugal', 
    'Slovakia', 'Slovenia', 'Spain'
]

euro_area_values = df_Irate.loc[
    'Euro area (EA11-1999, EA12-2001, EA13-2007, EA15-2008, EA16-2009, EA17-2011, EA18-2014, EA19-2015, EA20-2023)',
    '1994': '2024'
]


for country in eurozone_countries:
    df_Irate.loc[country, '1994': '2024'] = euro_area_values.values
```

```python
df_Irate = df_Irate.reset_index().melt(id_vars = 'TIME', var_name = 'Year', value_name = 'Irate')
```

```python
df_Irate = df_Irate.rename(columns = {'TIME':'Country'})
```

```python
df_Irate.sort_values(['Country', 'Year'], inplace = True)
```

```python
df_Irate.set_index(['Country', 'Year'], inplace = True)
```

```python
df_Irate.to_csv('df_Irate.csv')
```
