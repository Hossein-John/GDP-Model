# Concatenate all data

```python
import pandas as pd
```

```python
df_ED = pd.read_csv("df_ED.csv", index_col=['Country', 'Year'])
df_Exp = pd.read_csv("df_Exp.csv", index_col=['Country', 'Year'])
df_Irate = pd.read_csv("df_Irate.csv", index_col=['Country', 'Year'])
df_population = pd.read_csv("df_population.csv", index_col=['Country', 'Year'])
df_taxes = pd.read_csv("df_taxes.csv", index_col=['Country', 'Year'])
df_trade = pd.read_csv("df_trade.csv", index_col=['Country', 'Year'])
sectoral_GDP = pd.read_csv("sectoral_GDP.csv", index_col=['Country', 'Year'])
sectoral_Inv = pd.read_csv("sectoral_Inv.csv", index_col=['Country', 'Year'])

df_combined = pd.concat(
    [df_ED, df_Exp, df_Irate, df_population, df_taxes, df_trade, sectoral_GDP, sectoral_Inv], 
    axis=1
)
```
