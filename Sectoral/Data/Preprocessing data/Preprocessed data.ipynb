```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("combined_data.csv", na_values = ['NaN', ':'], index_col = ['Country', 'Year'])
```

```python
data
```

```python
missing_summary = data.isnull().sum()
```

```python
missing_summary 
```

```python
percentage_missing_1 = (data.isnull().mean() * 100).astype(int)
```

```python
percentage_missing_1
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (12,6))
percentage_missing_1.plot(kind = 'bar', color = 'red')
plt.title('Percentage of missing values in Each Column', fontsize = 14)
plt.ylabel('Missing Percentage', fontsize = 12)
plt.xlabel('Columns', fontsize = 12)
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```

# Keep year 1994 to 2023 ( 30 years )

```python
data_filtered = data[(data.index.get_level_values('Year') >= 1994) & (data.index.get_level_values('Year') < 2024)]
```

```python
data_filtered
```

```python
percentage_missing_2 = (data_filtered.isnull().mean() * 100).astype(int)
```

```python
percentage_missing_2
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (12,6))
percentage_missing_2.plot(kind = 'bar', color = 'red')
plt.title('Percentage of missing values in Each Column', fontsize = 14)
plt.ylabel('Missing Percentage', fontsize = 12)
plt.xlabel('Columns', fontsize = 12)
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```

```python
not_missed_columns = data_filtered.groupby(level = 0).apply(lambda x: (x.notnull().sum() > 0 ).sum())
```

```python
not_missed_columns.to_csv('not_missed_columns.csv')
```

```python
total_countries = data_filtered.index.get_level_values('Country').nunique()
countries_with_all_values = data_filtered.groupby(level=0).apply(lambda x: (x.notnull().sum() > 0).sum() == 14).sum()
print(f"Total number of countries: {total_countries}")
print(f"Number of countries with values in all 14 columns: {countries_with_all_values}")
```

# Countries to keep 

```python
countries_with_all_values_list = data_filtered.groupby(level=0).apply(lambda x: (x.notnull().sum() > 0).sum() == 14)

countries_with_all_values_list = countries_with_all_values_list[countries_with_all_values_list].index

filtered_data_1 = data_filtered.loc[countries_with_all_values_list]
```

```python
countries_with_all_values_list
```

```python
filtered_data_1
```

```python
percentage_missing_3 = (filtered_data_1.isnull().mean() * 100).astype(int)
```

```python
percentage_missing_3
```

```python
plt.figure(figsize = (12,6))
percentage_missing_3.plot(kind = 'bar', color = 'red')
plt.title('Percentage of missing values in Each Column', fontsize = 14)
plt.ylabel('Missing Percentage', fontsize = 12)
plt.xlabel('Columns', fontsize = 12)
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```

```python
Irate_missing = (filtered_data_1['Irate'].isnull().groupby(level = 0).mean() * 100).astype(int)
Irate_missing
```

# Countries to remove

```python
countries_to_remove = ["Bulgaria", "Czechia", "Hungary", "Poland", "Romania", "Sweden"]
filtered_data_1 = filtered_data_1[~filtered_data_1.index.get_level_values('Country').isin(countries_to_remove)]
```

```python
remaining_countries = filtered_data_1.index.get_level_values('Country').unique()
```

```python
print(f"Remaining countries after removal: {len(remaining_countries)}")
print(remaining_countries)
```

```python
percentage_missing_4 = (filtered_data_1.isnull().mean() * 100).astype(int)
```

```python
percentage_missing_4
```

```python
plt.figure(figsize = (12,6))
percentage_missing_4.plot(kind = 'bar', color = 'red')
plt.title('Percentage of missing values in Each Column', fontsize = 14)
plt.ylabel('Missing Percentage', fontsize = 12)
plt.xlabel('Columns', fontsize = 12)
plt.xticks(rotation = 45, ha = 'right')
plt.savefig("missing_values_plot.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
filtered_data_1.to_csv('countries_with_all_values(30 years).csv')
```
